```python
# Below values are string of list
# and we have to convert sring of list to list using 'ast' module

# [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]

import ast
```

```python
# using 'ast' module to convert dictionary into name

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
```

```python
movies_df.head()
```

```python
def convert(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert)
```

```python
movies_df.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
    
```

```python
movies_df.loc[:, 'crew'] = movies_df['crew'].apply(fetch_director)
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['cast'] = movies_df['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
movies_df['crew'] = movies_df['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
# concatenating the columns (means merging required columns into a single column)

movies_df['tags'] = movies_df['overview'] + movies_df['keywords'] + movies_df['genres'] + movies_df['crew'] + movies_df['cast']
```

```python
# converting list of tags column in string

new_df['tags'] = new_df['tags'].apply(lambda x: " ". join(x))
```

```python
# converting the sentences in tags into lower case

new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# creating helper function called stem

def stem(text):
    y = []
    
    for i in text.split():
         y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
ps.stem('zombies')
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d action adventure fantasy sciencefiction jamescameron samworthington zoesaldana sigourneyweaver')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features = 2000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
# to see the most common 2000 words that we want 

cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# passing vectors in cosine similarity function

similarity = cosine_similarity(vectors)
```

```python
# 1st movie will have the value 1 in first because its showing the similarity likewise 2nd movie will show similarity with 2nd words and so on.

similarity[0]
```

```python
# sorting in decesding order

sorted(similarity[0], reverse = True)
```

```python
# to intact the similarity between the movies we'll call enumerates function

list(enumerate(similarity[0]))

# about code shows that 0th movie distance with 0 movie is 1 
#                       0th movie distance with 1st movie is 0.1056 and so on.
```

```python
sorted(list(enumerate(similarity[0])), reverse = True)

# this code shows sorting on the basis of last no but we have to sort on the basis of 2nd no
```

```python
# sorting on the basis of 2nd no
# [1:6] taking 5 movies

sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
# creating function called recommend, (if anyone gives a movie it will recommend 5 similar movies)

def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```
